### Text2Text Generation Models | Seq2Seq Models | Encoder-Decoder Models

- Search the model according to task the model can perform =  https://huggingface.co/models
- In our problem statement i want to work with Text to Text generation task so iam going to import the model which it can perform the particular mention task and the model name is flan-t5-base which it is developed by google  https://huggingface.co/google/flan-t5-base 

```python
# creating a prompt
prompt = PromptTemplate(
    input_variables=["product"],
    template="What is a good name for a company that makes {product}"
)
```

```python
# Creating a chain
chain = LLMChain(llm=HuggingFaceHub(repo_id='google/flan-t5-large', model_kwargs={'temperature':0}),prompt = prompt) # Here in this code i have created a chain which takes the componets like hugging face text to text google flant t5 model and prompt template for generating the output on this taken prompt by the hugging face model and this chain funcationality is provided by langchain, and model_kwargs tells how creatly model should generate outout for the given prompt
```

```python
chain.run("colorful clothes") # just running the chain.run langchain hit huggingface one of the model to generate the output for the taken prompt and this is all achieved by chain concept
```

### Lets see how to create a pipeline which it downloads the Hugging face models into our local machine so we can use that downloaded model in our local machien

```python
# importing necessary libraries and modules for building a pipeline which it downloads the model in local system
from langchain.llms import HuggingFacePipeline
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, AutoModelForSeq2SeqLM
```

```python
model_id='google/flan-t5-large'  # here i have copied the model name from the hugging face application-> models-> search for any 1 model pasted the model name and saved in the variable name model_id which this model iam going to download in my local system memory to perform the text to text task by our local system model
```

```python
# if we want to hugging face pipeline 1st i need to perform the tokenization means before generating a output from downloaded model, first prompt should get tokenize which results in token words and then by using the autotokenizer we are going to encode the toknizer words into values 
```

```python
tokenizer=AutoTokenizer.from_pretrained(model_id) # here i have create the object for the class AutoTokenizer and passed the model_id so we get the encoded values of tokenizer words
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained(model_id,device_map='auto') # here i have create the object for the class AutoModelForSeq2SeqLM from which i calling the method called from_pre_trained to perform some operation like by passed the model_id and device_map arguments and finally after performing the operation model get saved in model object variable and operation is called downloading the model and saving in the model variable
```

```python
# creating pipeline 
pipeline = pipeline("text2text-generation", model=model, tokenizer=tokenizer, max_length=128) # here i have passed the key name which is text2text-generation and passed 2 arguments which are downloaded model and tokenizer and max_length tell max tokens to generate in the output
```

```python
# finally passing pipeline to the hugging face pipeline to perform the operations in sequence
local_llm = HuggingFacePipeline(pipeline=pipeline)
```

```python
# creating prompt
prompt = PromptTemplate(
    input_variables=["name"],
    template="Can you tell me about footballer {name}"
)
```

```python
# creating chain which consist of downloaded hugging face pre trained model which its going to tokenize the prompt words and encode it into values
chain = LLMChain(llm=local_llm,prompt = prompt)
```

```python
# running chain to generate the output for prompt throught langchain wrapper which it uses the downloaded huggingface model and this all achieved by langchain framework of chain functionality 
chain.run("sunil chetri") 
```
